```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('Survey1.csv', encoding='ISO-8859-1')
df2 = pd.read_csv('Survey2.csv', encoding='ISO-8859-1')
```

```python
df1['student_id'] = df1['Email address'].str.extract(r'^(\w+)@')
df2['student_id'] = df2['Email'].str.extract(r'^(\w+)@')
```

```python
common_students = pd.merge(df1, df2, on='student_id')
```

```python
count_survey1 = len(df1['student_id'].unique())
count_survey2 = len(df2['student_id'].unique())
count_common = len(common_students['student_id'].unique())
```

```python
labels = ['Survey 1', 'Survey 2', 'Common']
counts = [count_survey1, count_survey2, count_common]
```

```python
plt.bar(labels, counts, color=['blue', 'green', 'red'])
plt.xlabel('Surveys')
plt.ylabel('Number of Students')
plt.title('Number of Students in Each Survey and Common Students')
plt.show()
```
